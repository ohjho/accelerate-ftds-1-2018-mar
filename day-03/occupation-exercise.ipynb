# Exercise 5: what is the male to female ratio for each occupation?



```python
# what is the male to female ratio for each occupation?

# create a function that return 1 for male and 0 for female
def gender_to_numeric(x):
    if x == "M":
        return 1
    else:
        return 0

# create new column for the count

df['maleorfemale'] = df['gender'].apply(gender_to_numeric)  # we can use .apply to generate new columns for it

# create a temporarily dataframe and with percentage 
temp = df.groupby('occupation').maleorfemale.sum()/df.occupation.value_counts()*100


# sort to the most male 
temp.sort_values(ascending = False)

# that's nice to conclude with common sense of male domainating occupation
```

# Exercise 6: what is the maximun and minimum age for each occupation?

```python
# This is a more condense way to work out the solution

df.groupby('occupation').age.agg(['max','min','median']).sort_values(by = 'median') #with sorting, it is more meaningful
```

```python
# how about we further factor it with female and occupation?

df.groupby(['occupation','gender']).age.agg(['mean','min','max'])
```
