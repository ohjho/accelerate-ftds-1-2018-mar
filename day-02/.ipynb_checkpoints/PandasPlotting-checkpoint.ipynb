# Pandas

Pandas has plotting built in, it provides convient ways to quickly make plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.tools.plotting import andrews_curves
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv')
df.head()
```

# Bars

```python
df.species.value_counts().plot('bar')
```

```python
df.groupby('species').mean().plot.bar()
```

```python
df.groupby('species').mean().plot.bar(stacked = True)
```

# Box Plots

```python
df.plot.box()
```

# Histograms

```python
df.plot.hist()
```

# KDE

```python
df.plot.kde()
```

# Andrews Curves


```python
andrews_curves(df,'species')
```
